### Question 1: Extract MFCCs

```python
file_path = './reference.wav'

# Load the audio file
y, sr = librosa.load(file_path, sr=None)  # y is a numpy array, sr = sampling rate = 16k Hz

# Define frame sizes in ms
frame_sizes = [20, 30, 40]
frame_sizes_samples = [int(size * sr / 1000) for size in frame_sizes]  # Convert to samples: how many sampling points in a window

# Initialize a list to hold the MFCCs
mfccs_list = []

# Extract MFCCs for each frame size
for frame_size in frame_sizes_samples:
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13, n_fft=frame_size, hop_length=frame_size // 2)
    mfccs_list.append(mfccs)

# Plot the MFCCs of the 100th frame for each frame size
plt.figure(figsize=(12, 8))

for i, (frame_size, mfccs) in enumerate(zip(frame_sizes, mfccs_list)):
    if mfccs.shape[1] > 100:  # Ensure there are at least 100 frames
        plt.subplot(len(frame_sizes), 1, i + 1)
        plt.plot(mfccs[:, 99])  # 100th frame (index 99)
        plt.title(f'MFCCs of 100th Frame (Frame Size: {frame_size} ms)')
        plt.xlabel('MFCC Coefficients')
        plt.ylabel('Value')
        plt.grid()

plt.tight_layout()
plt.show()
```
